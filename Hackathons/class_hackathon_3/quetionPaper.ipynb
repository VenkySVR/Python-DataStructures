```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel(r"CurrencyDataFile.xlsx")
df
```

```python
q=open("questions.txt","a")
ans=open("answers.txt","a")


def selectque():
    q_no=np.random.randint(1,(len(df)-1))
    return q_no
```

```python
def quetion():
    op=np.random.choice([1,2,3])
    if op==1:
        q.write("what is the symbol of "+df.loc[q_no][0]+" ?\n")
        q.write("A: "+df.loc[q_no][1]+"\tB: "+df.loc[q_no+1][1]+"\tC: "+df.loc[q_no-1][1]+"\n\n")
        ans.write("A: "+df.loc[q_no][1]+"\n")
    elif op==2:
        q.write("what is the symbol of "+df.loc[q_no][0]+" ?\n")
        q.write("A: "+df.loc[q_no+1][1]+"\tB: "+df.loc[q_no][1]+"\tC: "+df.loc[q_no-1][1]+"\n\n")
        ans.write("B: "+df.loc[q_no][1]+"\n")
    elif op==3:
        q.write("what is the symbol of "+df.loc[q_no][0]+" ?\n")
        q.write("A: "+df.loc[q_no-1][1]+"\tB: "+df.loc[q_no+1][1]+"\tC: "+df.loc[q_no][1]+"\n\n")
        ans.write("C: "+df.loc[q_no][1]+"\n")
```

```python
no_q_papers=int(input("enter number of quetion papers:\t"))
no_quetions = int(input("enter number of quetions:\t"))
paper = 1
all_q=[]
for q_paper in range(0,no_q_papers):
    quetion_number=1
    q.write("****** Quetion paper : "+str(paper)+" ********\n\n")
    ans.write("****** Quetion paper : "+str(paper)+" key ********\n\n")
    paper+=1
    if q_paper ==0:
        for que in range(0,no_quetions):
            q_no=selectque()
            all_q.append(q_no)
            q.write("Q)  "+str(quetion_number)+"\n")
            ans.write("Q)  "+str(quetion_number)+"\t")            
            quetion_number+=1
            quetion()
    else:
        np.random.shuffle(all_q)
        for q_no in all_q:
            q.write("Q)  "+str(quetion_number)+"\n")
            ans.write("Q)  "+str(quetion_number)+"\t")
            quetion_number+=1
            quetion()
    q.write("\n\n")
    ans.write("\n\n")
q.close()
ans.close()
```
