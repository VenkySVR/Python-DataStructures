```python
import pandas as pd
import requests
import matplotlib.pyplot as plt


data = requests.request("GET", 'https://pomber.github.io/covid19/timeseries.json').json()
```

```python
timedataframe = {}    
```

```python
for name in data:
        for j in range(0, len(data[name])):
            if data[name][j]['confirmed'] >= 1:
                break
        for i in range(0, len(data[name])):
            if data[name][i]['deaths'] >= 1:
                death = i-j+1
                break
        for k in range(0, len(data[name])):
            if data[name][k]['confirmed'] >= 1:
                break
        for l in range(0, len(data[name])):
            if data[name][l]['recovered'] >= 1:
                rec = l-k+1
                break
        timedataframe[name]={
        "confirmed-recovery days":rec,
        "confirmed-death days":death
            }
```

```python
timedataframe=pd.DataFrame(timedataframe)
timedataframe
```

```python
df=timedataframe.T
df
```
