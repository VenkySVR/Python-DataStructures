```python
import pandas as pd
import matplotlib.pyplot as plt
```

# the death growing rate is more than 2% of the confirmed cases at any point of time, identify the country as a HIGH RISK COUNTRY. 

```python
df1 = pd.read_csv("time_series_covid19_confirmed_global.csv")
df1=df1.set_index('Country/Region')
df1
```

```python
df2 = pd.read_csv('time_series_covid19_deaths_global.csv')
df2=df2.set_index('Country/Region')
df2
```

```python
df=df2.div(df1)*100
df=df
```

```python
df=df.fillna(0)
```

```python
df
```

```python
for i in range (0,len(df.index),10):
    df.iloc[i:i+10].T.plot()
```
